```python
recent_video = {}
from googleapiclient.discovery import build

# Cấu hình thông tin API key và tạo kết nối với YouTube Data API
api_key = api_key
youtube = build('youtube', 'v3', developerKey=api_key)

# Danh sách các IDs của các kênh
channel_ids = ids
# Duyệt qua từng ID kênh
for channel_id in channel_ids:
    try:
        recent_video[channel_id] = list()
        # Gọi API để lấy thông tin về playlist chứa các video mới nhất của kênh
        playlist_response = youtube.channels().list(
            part='contentDetails',
            id=channel_id
        ).execute()
        
        # Lấy ID của playlist chứa các video mới nhất
        playlist_id = playlist_response['items'][0]['contentDetails']['relatedPlaylists']['uploads']
        
        # Gọi API để lấy danh sách các video trong playlist
        playlist_items_response = youtube.playlistItems().list(
            part='snippet',
            playlistId=playlist_id,
            maxResults=10
        ).execute()
        
        # Duyệt qua từng video và lấy ID của video
        for video in playlist_items_response['items']:
            video_id = video['snippet']['resourceId']['videoId']
            recent_video[channel_id].append(video_id)
    
    except Exception as e:
        print(f"Error occurred for channel ID {channel_id}: {str(e)}")
```

```python
recent_video_df =  pd.DataFrame({"chanel_ids" : recent_video.keys(), "video_ids": recent_video.values()}) 
recent_video_df
```
